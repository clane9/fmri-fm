```python
from pathlib import Path
import numpy as np
from sklearn.decomposition import PCA

import datasets as hfds
from tqdm import tqdm
```

```python
input_spaces = ["schaefer400", "flat", "mni_cortex"]

for space in input_spaces:
    outpath = Path(f"results/pca_{space}.npz")
    # if outpath.exists():
    #     print(f"{outpath} exists; skipping")

    url = f"s3://medarc/fmri-datasets/eval/hcpya-clips.{space}.arrow/validation"
    dataset = hfds.load_dataset("arrow", data_files=f"{url}/*.arrow", split="train")
    dataset.set_format("numpy")
    print(dataset)

    X = []
    for sample in tqdm(dataset):
        X.append(sample["bold"][:1])
    X = np.concatenate(X)
    print(X.shape)

    X_norm = (X - X.mean(axis=1, keepdims=True)) / X.std(axis=1, keepdims=True)
    pca = PCA(n_components=8, svd_solver="randomized").fit(X_norm)
    components = pca.components_
    sign = np.sign(components.sum(axis=1, keepdims=True))
    components = components * sign
    print(components.shape)

    np.savez(outpath, components=components)
```
